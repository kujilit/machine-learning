```python
import pandas as pd

df = pd.read_csv('Boston.csv')

df.head()
```

```python
import statsmodels.formula.api as sm

model = sm.ols('medv ~ dis', df, subset=None, drop_cols=None).fit()
print(model.summary())
```

```python
import seaborn as sb

sb.regplot(data=df, x="medv", y="dis")
```

```python
model = sm.ols('medv ~ dis + crim + zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + black + lstat', df, subset=None, drop_cols=None).fit()
print(model.summary())
```
